```python
import os
from rdflib import Graph
import pandas as pd

# 1. Chemins pr√©cis
dossier_source = r"C:\Python_challenges\Brief_jury_blanc\rdf_CTA"
dossier_dest = r"c:\Python_challenges\Brief_jury_blanc\output"

# 2. Cr√©ation du dossier
if not os.path.exists(dossier_dest):
    os.makedirs(dossier_dest)
    print(f" Dossier cr√©√© avec succ√®s : {dossier_dest}")
else:
    print(f" Le dossier existe d√©j√† : {dossier_dest}")

# 3. Boucle de traitement
for fichier in os.listdir(dossier_source):
    if fichier.endswith(".rdf"):
        print(f"‚è≥ Conversion de {fichier}...")
        
        chemin_rdf = os.path.join(dossier_source, fichier)
        nom_csv = fichier.replace(".rdf", "_propre.csv")
        chemin_csv = os.path.join(dossier_dest, nom_csv)
        
        try:
            g = Graph()
            g.parse(chemin_rdf, format="xml")
            
            # Transformation des triplets en colonnes
            data = []
            for s, p, o in g:
                # On nettoie le nom de la colonne
                nom_col = str(p).split('/')[-1].split('#')[-1]
                data.append({'id': str(s), 'propriete': nom_col, 'valeur': str(o)})
            
            df_raw = pd.DataFrame(data)
            # On pivote pour avoir 1 ligne par ID et des colonnes propres
            df_pivot = df_raw.pivot_table(index='id', 
                                          columns='propriete', 
                                          values='valeur', 
                                          aggfunc='first').reset_index()
            
            df_pivot.to_csv(chemin_csv, index=False)
            
        except Exception as e:
            print(f"‚ùå Erreur sur {fichier} : {e}")

print("\n‚ú® Tous les fichiers ont √©t√© convertis dans le dossier 'csv_resultats' !")
```

```python
import os
import pandas as pd

# --- CONFIGURATION DES CHEMINS ---
# Le dossier o√π sont vos 19 fichiers CSV convertis
dossier_source = r"C:\Python_challenges\Brief_jury_blanc\output"

# Le NOUVEAU dossier pour la fusion (Chemin corrig√© avec votre email)
dossier_fusion = r"C:\Python_challenges\Brief_jury_blanc\fusion_finale"

nom_fichier_final = "CTA_Ridership_Total_Fusionne.csv"

# --- EXECUTION ---

# 1. Cr√©ation du dossier propre
if not os.path.exists(dossier_fusion):
    os.makedirs(dossier_fusion)
    print(f"‚úÖ Dossier cr√©√© : {dossier_fusion}")

# 2. Fusion des fichiers
liste_df = []

for fichier in os.listdir(dossier_source):
    if fichier.endswith("_propre.csv"):
        chemin_complet = os.path.join(dossier_source, fichier)
        df_temp = pd.read_csv(chemin_complet)
        liste_df.append(df_temp)
        print(f"‚ûï Ajout de : {fichier}")

if liste_df:
    # On rassemble tout
    df_global = pd.concat(liste_df, ignore_index=True)
    
    # Chemin complet du fichier final
    chemin_final_csv = os.path.join(dossier_fusion, nom_fichier_final)
    
    # Sauvegarde
    df_global.to_csv(chemin_final_csv, index=False)
    
    print("-" * 30)
    print(f"‚ú® FUSION R√âUSSIE !")
    print(f"üìä Nombre total de lignes : {len(df_global)}")
    print(f"üìç Fichier sauvegard√© dans : {dossier_fusion}")
else:
    print("‚ùå Aucun fichier trouv√© dans le dossier source.")
```

```python
import pandas as pd

# Chemin de votre fichier fusionn√©
chemin_fichier = r"C:\Python_challenges\Brief_jury_blanc\fusion_finale\CTA_Ridership_Total_Fusionne.csv"
try:
    # 1. Chargement (low_memory=False √©vite les erreurs sur les gros fichiers)
    df = pd.read_csv(chemin_fichier, low_memory=False)
    
    # 2. Suppression des colonnes inutiles
    colonnes_a_supprimer = ['id', 'type', 'member', 'rowID']
    df = df.drop(columns=[c for c in colonnes_a_supprimer if c in df.columns])
    
    # 3. Conversion de la colonne date
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    
    # 4. Filtrage des ann√©es 2019 √† 2025
    # On garde les lignes o√π l'ann√©e est comprise entre 2019 et 2025
    df = df[(df['date'].dt.year >= 2019) & (df['date'].dt.year <= 2025)]
    
    # 5. On remet la date au format texte propre (AAAA-MM-JJ) pour le CSV
    df['date'] = df['date'].dt.strftime('%Y-%m-%d')
    
    # 6. On sauvegarde dans le M√äME fichier pour le mettre √† jour
    df.to_csv(chemin_fichier, index=False)
    
    print(f" Nettoyage termin√© !")
    print(f" Nouveau total de lignes : {len(df)}")
    print(f" Le fichier a √©t√© mis √† jour : {chemin_fichier}")
    
    # Affichage des 10 premi√®res lignes pour v√©rifier
    display(df.head(10))

except Exception as e:
    print(f" Erreur : {e}")
```
